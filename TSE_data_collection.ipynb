## 東証一部上場気企業の株価データを Quandl から収集
以下のコードを実行する。

ただし，実行に２時間程度を要するため，気長に待ちましょう。

また，うまくデータを取得できたかは（コード中の）変数 path で指定したディレクトリを開けば分かります。

```python
# Quandl から東証一部上場企業のデータを収集するコード

import pandas as pd
import quandl
import sys

quandl.ApiConfig.api_key = "swWCyaT7PznGyDJwoPys"

# (TODO) TSE_corp_list.csv のパスは各自の環境に合わせて変更すること
companies = pd.read_csv("/content/drive/My Drive/BASE2/EconomicData/TSE_corp_list.csv")

for code in companies['コード']:
    # print(code)
    indicator = "TSE/{}".format(code)
    path = "/content/drive/My Drive/TSEdata/{}.csv".format(code)
    
    # ファイルが既に存在する場合はスキップ
    if os.path.exists(path):
        print("data {} exist".format(indicator), file=sys.stderr)
    else:
        print("data {} collect ...".format(indicator), file=sys.stderr)
        data = quandl.get(indicator)
        data.to_csv(path) # CSV ファイルへの保存
```
